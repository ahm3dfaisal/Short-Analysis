```python
  #Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from wordcloud import WordCloud,STOPWORDS
import plotly.express as px
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
fig = pd.crosstab(data["Gender"],data["satisfaction"]).plot(kind="bar",
                                                             figsize = (12,7),
                                                             linewidth = 1,
                                                             edgecolor = "w"*2)
fig.set_title("Gender vs Satisfaction Levels")
plt.show()
```

```python
bins = [0,5,15,30,60,120,180,500]
labels=['Less than 5 min', '5-14 min','15-29 min', '30-59 min', '60-119 min', '120-179 min', 'More than 3 hours']
```

```python
data['Time ranges(Departure Delay)'] = pd.cut(data['Departure Delay in Minutes'], bins, labels=labels, include_lowest = True)
```

```python
data['Time ranges(Arrival Delay)'] = pd.cut(data['Arrival Delay in Minutes'], bins, labels=labels, include_lowest = True)
```

```python
data.head(25)
```

```python
plt.figure(figsize=(7,7))
work_hours = data["Time ranges(Departure Delay)"].value_counts().reset_index()

ax = sns.barplot("Time ranges(Departure Delay)","index",data=work_hours[:20],linewidth=2,edgecolor="k"*20)
plt.xlabel("Number of Customers")
plt.ylabel("Delay")
plt.title("Delay experienced by customers (Departure)")
plt.grid(True,alpha=.3)
```

```python
plt.figure(figsize=(7,7))
work_hours = data["Time ranges(Arrival Delay)"].value_counts().reset_index()

ax = sns.barplot("Time ranges(Arrival Delay)","index",data=work_hours[:20],linewidth=2,edgecolor="k"*20)
plt.xlabel("Number of Customers")
plt.ylabel("Delay")
plt.title("Delay experienced by customers (Arrival)")
plt.grid(True,alpha=.3)
```

```python
bins2 = [0,1000,2000,3000,4000]
labels2 = ['Less than 1000 miles', '1000-1999 miles', '2000-2999 miles', '3000-3999 miles']
```

```python
data['Flight distance range'] = pd.cut(data['Flight Distance'], bins2, labels=labels2, include_lowest = True)
```

```python
data.head(25)
```

```python
fig = pd.crosstab(data["Flight distance range"],data["satisfaction"]).plot(kind="barh",
                                                             figsize = (7,7),
                                                             linewidth = 1,
                                                             edgecolor = "w"*2)
fig.set_title("Flight Distance Vs Satisfaction")
plt.show()
```
